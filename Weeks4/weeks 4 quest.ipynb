## 4. 다음을 만족하는 20년 달력 데이터 프레임을 만드시오.  
|index||original||Year||Month||Day||weekday||weekend||season|
|---||---||---||---||---||---||---||---|
|0||20-01-01||20||1||1||Wed||0||winter|  
|1||20-01-02||20||1||2||Thr||0||winter|  
|2||20-01-03||20||1||3||Fri||0||winter| 
|3||20-01-03||20||1||4||Sat||1||winter|  
|...||...||...||...||...||...||...||...|    
|365||20-12-31||20||12||31||Thr||0||winter|  

```python
import datetime
four_df = pd.DataFrame()
# original 날짜
four_df["original"] = pd.date_range("2020-01-01", periods= 366)
# year
four_df["Year"] = four_df["original"].dt.year
# Month
four_df["Month"] = four_df["original"].dt.month
# Day
four_df["Day"] = four_df["original"].dt.day
# weekday
four_df["Weekday"] = four_df["original"].dt.day_name()
# weekend
# 주말 = 1 , 주중 = 0
weekend = []
for i in range(0,366) : # 365까지 
    if int(four_df["original"].dt.weekday[i]) >= 6  : # day 6, 7 = 금, 토
        weekend.append(1)
    else : 
        weekend.append(0)
four_df["Weekend"] = weekend
# season
season = []
for i in range(0,366) :
    if four_df["original"].dt.month[i] == 3 or four_df["original"].dt.month[i] == 4 or four_df["original"].dt.month[i] == 5 :
        season.append("Spring")
    elif four_df["original"].dt.month[i] == 6 or four_df["original"].dt.month[i] == 7 or four_df["original"].dt.month[i] == 8 :
        season.append("Summer")
    elif four_df["original"].dt.month[i] == 9 or four_df["original"].dt.month[i] == 10 or four_df["original"].dt.month[i] == 11 :
        season.append("Autumn")
    else : 
        season.append("Winter")
four_df["Season"] = season
four_df
```
